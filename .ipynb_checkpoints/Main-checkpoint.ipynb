## 使用mask替换主体

```python
copydf = copy.deepcopy(webquestion_trainmodel)
copydf.head()
```

```python
be_list = ['is','was','am', 'is', 'are','was','were','be']
do_list = ['do','does','did','done','doing']
def modify(inputStr):
    islast = True
    inputStr = inputStr.strip().lower()
    if(inputStr[-1]=='?'):
        inputStr = inputStr[:-1]
    tokens = nltk.word_tokenize(inputStr)
    tagged = nltk.pos_tag(tokens)
    
    verb_index = 0;
    verb_word = '';

    
    #处理并删除起始疑问词
    if((tagged[0][0]=='where')|((tagged[0][0]=='when'))):
        tagged.pop(0)
        tagged.append(('in','IN'))
        tagged.append(('[MASK]','NN'))
        return ' '.join([t[0] for t in tagged])+'.'
    elif((tagged[0][0] == 'who')|(tagged[0][0] == 'whom')):
        tagged.pop(0)
        tagged.insert(0,('[MASK]','NN'))
        return ' '.join([t[0] for t in tagged])+'.'
    elif((tagged[0][0] == 'which')|(tagged[0][0] == 'what')):
        tagged.pop(0)
    elif(tagged[0][0] == 'how'):
        tagged.pop(0)
        tagged.insert(len(tagged),('by','IN'))
        
        
    #恢复陈述句语序
    for index in range(len(tagged)):
        if(tagged[index][0] in be_list):
            verb_index = index
            verb_word = tagged[index]
            tagged = tagged[verb_index+1:]
            #判断最后是不是介词
            if(tagged[len(tagged)-1][1] not in ['IN','TO']):
                tagged.append(verb_word)
            tagged.append(('[MASK]','NN'))
            return ' '.join([t[0] for t in tagged])+'.'
        if((tagged[index][0] in do_list)|(tagged[index][1]=='MD')):
            verb_index = index
            verb_word = tagged[index][0]
            tagged = tagged[verb_index+1:]
            tagged.append(('[MASK]','NN'))
            return ' '.join([t[0] for t in tagged])+'.'
    

    
    
    if islast:
        tagged.append(('[MASK]','NN'))
    return ' '.join([t[0] for t in tagged])+'.'
```

```python
##测试问句转换陈述句效果
question_one = 'when has been offered to the poor homeless children for the coming winter?'
question_two = 'who movies does taylor lautner play in?'
question_three = 'what country is russia close to?'

q1 = modify(question_one)
print(q1)
q2 = modify(question_two)
print(q2)
q3 = modify(question_three)
print(q3)
```

```python
for index, row in copydf.iterrows():
    row[0] = modify(row[0])
```
